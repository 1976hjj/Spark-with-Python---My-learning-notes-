```
df_people_flow = spark.read.option("inferSchema","true").option("header","true").csv("/FileStore/tables/global_flow_of_people.csv")
```

```
df_people_flow.printSchema()
```

```
from pyspark.sql import Window
from pyspark.sql.functions import rank, min
window = Window.partitionBy("country_orig").orderdBy("regionflow_1990").rowsBetween(-1, 1)
df_people_flow.select(rank().over(window), min("regionflow_1990").over(window))
```
