```
list_A = [{'name':'A','ID':10},{'name':'B','ID':20}]
list_B = [{'ID':10,'ADDR':'DE'},{'ID':30,'ADDR':'IND'}]
df_A = spark.createDataFrame(list_A)
df_B = spark.createDataFrame(list_B)
```

```
df_A.crossJoin(df_B).show()
df_A.crossJoin(df_B.select("ADDR")).select("ID","name","ADDR").show()
```

```
#explain(True)
#join : inner, cross, outer, full, full_outer, left, left_outer, right, right_outer, left_semi, and left_anti
df_A.join(df_B,df_A.ID == df_B.ID, 'inner').show()
df_A.join(df_B,df_A.ID == df_B.ID, 'leftouter').show()
df_A.join(df_B,df_A.ID == df_B.ID, 'leftouter').explain(True)
```

```
#hint
df_A.join(df_B.hint("broadcast"),"ID").explain()
df_A.join(df_B,df_A.ID == df_B.ID, 'inner').explain()
```

```
df_A.intersect(df_B).show()
```

```
#repartition
df_res = df_A.crossJoin(df_B.select("ADDR")).select("ID","name","ADDR")
df_res.show()
df_res.repartition(2,"ADDR").show()
df_res.repartition("ADDR").show()
```

```
#crosstab
df_A.crosstab("ID","name").show()
df_B.crosstab("ID","ADDR").show()
```

```
#cube & rollup
#https://stackoverflow.com/questions/37975227/what-is-the-difference-between-cube-and-groupby-operators
df_A.cube("ID","name").count().show()
df_A.rollup("ID","name").count().show()
```

```
df_A.cube("ID","name").count().show()
df_A.cube("ID","name").count().fillna(100).show()
```
