```

df = spark.createDataFrame([{'name':'vivek B','id':100},{'name':'poll S','id':200},{'name':'peter M','id':150}])
df.show()
```

```
#Structs
from pyspark.sql.functions import struct, col
df_struct = df.select(struct("id","name").alias("struct"))
df_struct.show()
```

```
df_struct.createOrReplaceTempView("vw_struct")
df_struct.select("struct.name").show()
df_struct.select(col("struct").getField("id")).show()
```

```
df_struct.select("struct.*").show()
```

```
#Arrays
from pyspark.sql.functions import split, size, array_contains
df.select(split("name", " ").alias("array_name")).show()
df.select(split("name", " ").alias("array_name")).selectExpr("array_name[1]").show()
df.select(size(split(col("name"), " "))).show()
```

```
df.select(array_contains(split(col("name"), " "), "vivek")).show()
```

```
#Explode
from pyspark.sql.functions import explode
df.withColumn("splited" , split("name", " ")).withColumn("exploded", explode("splited")).select("ID", "exploded").show()
```
